```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cifar10_1_5000.csv")
```

```python
df.head()
```

```python
def visualize_csv(df: str) -> None:
    # Existing scatter plot
    plt.figure(figsize=(20, 20))
    fig, ax = plt.subplots()
    # name : remove .csv from df
    name = df[:-4]
    df = pd.read_csv(df)
    # Connect the points
    ax.plot(df["name"], df["test gm"], "r-")  # Change color here

    ax.scatter(
        df["name"], df["test gm"], color="red", label="bACC"
    )  # Change color here

    # Loop over the data points and add a text label for each point
    # for i in range(len(df["name"])):
    # ax.text(df["name"][i], df["test bal acc"][i], df["name"][i])
    ax.scatter(
        df["name"], df["test bal acc"], color="green", label="GM"
    )  # Change color here
    # for i in range(len(df["name"])):
    #  ax.text(df["name"][i], df["test gm"][i], df["name"][i])
    # Connect the points
    ax.plot(df["name"], df["test bal acc"], "g-")  # Change color here
    ax.legend()
    plt.xticks(rotation=45)
    plt.grid()
    # save the plot
    file_name = "./output/" + name + ".png"
    plt.savefig(file_name)
```

```python
# walk through the data in this directory
# First, make folders for the output
import os
import sys
import matplotlib.pyplot as plt
import pandas as pd

# get the current working directory
cwd = os.getcwd()
# get the list of files in the directory
files = os.listdir(cwd)
# filter the files to only include csv files
csv_files = [f for f in files if f.endswith(".csv")]
# print the list of csv files
print(csv_files)
```

```python
for f in csv_files:
    visualize_csv(f)
    print(f"Visualized {f}")
```
